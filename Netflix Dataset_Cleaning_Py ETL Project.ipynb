# Removing Nulls

```python
def remove_nulls(df,columns_null): 
    df_cleaned = df.dropna(subset=columns_null).copy()
    return df_cleaned
    
```

```python
columns_null = ['title','genres','releaseYear','imdbId','imdbAverageRating','imdbNumVotes']
df_cleaned = remove_nulls(df,columns_null)
```

```python
df_cleaned.isnull().sum()
```

```python
len(df_cleaned)
```

# Replace Nulls

```python
df_cleaned['url'] = df_cleaned['url'].fillna("Unknown")
```

```python
df_cleaned.isnull().sum()
```

# Convert from decimal to integer

```python
df_cleaned.info()
```

```python
def convert_to_integer(df,columns_float):
    for i in columns_float:
        df[i]=df[i].astype(int)
    return df
columns_float = ['releaseYear','imdbNumVotes']
```

```python
df_cleaned = convert_to_integer(df_cleaned,columns_float)
```

```python
df_cleaned.info()
```

# Convert From Object to String

```python
def convert_to_string(df,columns_object):
    for i in columns_object:
        df[i] = df[i].astype(str)
    return df
columns_object = ['url','title','type','genres','imdbId','availableCountries']
```

```python
df_cleaned = convert_to_string(df_cleaned,columns_object)
```

```python
df_cleaned.info()
```

```python
df_cleaned['type'] = df_cleaned['type'].str.title()
```
